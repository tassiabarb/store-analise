### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Faturamento total
total_loja = sum(loja.Preço)
total_loja2 = sum(loja2.Preço)
total_loja3 = sum(loja3.Preço)
total_loja4 = sum(loja4.Preço)
total = sum([total_loja, total_loja2, total_loja3, total_loja4])
print('Faturamento total ' + str(total))
```

```python
# Faturamento de cada loja separadamente
fat1 = loja['Preço'].sum()
fat2 = loja2['Preço'].sum()
fat3 = loja3['Preço'].sum()
fat4 = loja4['Preço'].sum()

print("Loja 1:", fat1)
print("Loja 2:", fat2)
print("Loja 3:", fat3)
print("Loja 4:", fat4)

```

# 2. Vendas por Categoria

```python
# Loja 1
cat1 = loja['Categoria do Produto'].value_counts()
print("Vendas por categoria - Loja 1")
print(cat1)
print("\n")

# Loja 2
cat2 = loja2['Categoria do Produto'].value_counts()
print("Vendas por categoria - Loja 2")
print(cat2)
print("\n")

# Loja 3
cat3 = loja3['Categoria do Produto'].value_counts()
print("Vendas por categoria - Loja 3")
print(cat3)
print("\n")

# Loja 4
cat4 = loja4['Categoria do Produto'].value_counts()
print("Vendas por categoria - Loja 4")
print(cat4)
```

# 3. Média de Avaliação das Lojas


```python
# Loja 1
aval1 = loja['Avaliação da compra'].mean()
print("Média de avaliação - Loja 1:", aval1)

# Loja 2
aval2 = loja2['Avaliação da compra'].mean()
print("Média de avaliação - Loja 2:", aval2)

# Loja 3
aval3 = loja3['Avaliação da compra'].mean()
print("Média de avaliação - Loja 3:", aval3)

# Loja 4
aval4 = loja4['Avaliação da compra'].mean()
print("Média de avaliação - Loja 4:", aval4)
```

# 4. Produtos Mais e Menos Vendidos

```python
print("LOJA 1 – Produtos mais vendidos:")
print(loja['Produto'].value_counts().head(5))
print("\n")

print("LOJA 1 – Produtos menos vendidos:")
print(loja['Produto'].value_counts().tail(5))
print("\n")
```

```python
print("LOJA 3 – Produtos mais vendidos:")
print(loja3['Produto'].value_counts().head(5))
print("\n")

print("LOJA 3 – Produtos menos vendidos:")
print(loja3['Produto'].value_counts().tail(5))
print("\n")
```

```python
print("LOJA 4 – Produtos mais vendidos:")
print(loja4['Produto'].value_counts().head(5))
print("\n")

print("LOJA 4 – Produtos menos vendidos:")
print(loja4['Produto'].value_counts().tail(5))
print("\n")
```

# 5. Frete Médio por Loja

```python
# Loja 1
frete1 = loja['Frete'].mean()
print("Custo médio de frete - Loja 1:", frete1)

# Loja 2
frete2 = loja2['Frete'].mean()
print("Custo médio de frete - Loja 2:", frete2)

# Loja 3
frete3 = loja3['Frete'].mean()
print("Custo médio de frete - Loja 3:", frete3)

# Loja 4
frete4 = loja4['Frete'].mean()
print("Custo médio de frete - Loja 4:", frete4)
```

```python
# Faturamento total por loja
import matplotlib.pyplot as plt

faturamentos = [fat1, fat2, fat3, fat4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos)
plt.title("Faturamento Total por Loja")
plt.xlabel("Loja")
plt.ylabel("Faturamento (R$)")
plt.show()
```

```python
# Distribuição das Categorias
plt.figure(figsize=(8,8))
plt.pie(cat1, labels=cat1.index, autopct='%1.1f%%')
plt.title("Distribuição das Categorias - Loja 1")
plt.show()
```

```python
# Média de Avaliação por Loja
avaliacoes = [aval1, aval2, aval3, aval4]

plt.figure(figsize=(8,5))
plt.plot(lojas, avaliacoes, marker='o')
plt.title("Média das Avaliações por Loja")
plt.xlabel("Loja")
plt.ylabel("Média de Avaliação")
plt.ylim(0, 5)
plt.grid(True)
plt.show()
```
